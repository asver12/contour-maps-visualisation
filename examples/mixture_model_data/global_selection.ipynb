```python
dataset = []
for i in range(2,6):
    models, xlabels, ylabels = MixtureModel.generate_gaussian_plots(model_mcg.data, ["year", "mpg_city","mpg_highway","displacement"], n_components=i)
    models_1, xlabels_1, ylabels_1 = MixtureModel.generate_gaussian_plots(model_allbus.data, ["age", "education","income","happiness","health"], n_components=i)
    models.extend(models_1)
    xlabels.extend(xlabels_1)
    ylabels.extend(ylabels_1)
    dataset.append([models, xlabels, ylabels])
```

```python
indices = {}
for ind, label in enumerate(zip(dataset[0][1],dataset[0][2])):
    indices["{}_{}".format(label[1],label[0])]=ind
    indices["{}_{}".format(label[0],label[1])]=ind
```

```python
def plot_img_at_inds(dataset, inds, *args, **kwargs):
    for i in inds:
            picture_plot.plot_image_variations(dataset[0][i], xlabels=dataset[1][i], ylabels=dataset[2][i], *args, **kwargs)
```

```python
def plot_img_at_inds_one(dataset, inds, *args, **kwargs):
    for i in inds:
            picture_plot.plot_images([dataset[0][i]], xlabels=dataset[1][i], ylabels=dataset[2][i], *args, **kwargs)
```

```python
def get_indices(values):
    return [indices[i] for i in values]
```

```python
img_inds = get_indices(["displacement_mpg_highway","age_income"])
```
