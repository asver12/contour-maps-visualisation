Simple test of the porter-duff-source-over operator

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import itertools
```

```python
from src import color_blending_operator
```

```python
points_1 = list(itertools.permutations(range(0,11),2))
points_2 = list(itertools.permutations(range(5,16),2))
points = list(set(points_1+points_2))
```

```python
rot = np.array([1,0,0],dtype=float)
gruen =  np.array([0,1,0],dtype=float)
```

```python
alpha = 0.5
mixed = color_blending_operator.porter_duff_source_over(rot,gruen,alpha)
```

```python
colors = []
for i in points:
    if i[0] > 4 and i[1] > 4 and i[0] < 11 and i[1] < 11:
        colors.append(mixed)
    elif i[0]<=4 or i[1] <= 4:
        colors.append(rot)
    else:
        colors.append(gruen)
```

```python
plt.scatter(*zip(*points),c=colors)
plt.show()
```

```python
alpha = 0.2
mixed = color_blending_operator.porter_duff_source_over(rot,gruen,alpha)
colors = []
for i in points:
    if i[0] > 4 and i[1] > 4 and i[0] < 11 and i[1] < 11:
        colors.append(mixed)
    elif i[0]<=4 or i[1] <= 4:
        colors.append(rot)
    else:
        colors.append(gruen)
plt.scatter(*zip(*points),c=colors)
plt.show()
```

```python
alpha = 0.8
mixed = color_blending_operator.porter_duff_source_over(rot,gruen,alpha)
colors = []
for i in points:
    if i[0] > 4 and i[1] > 4 and i[0] < 11 and i[1] < 11:
        colors.append(mixed)
    elif i[0]<=4 or i[1] <= 4:
        colors.append(rot)
    else:
        colors.append(gruen)
plt.scatter(*zip(*points),c=colors)
plt.show()
```

```python
points.sort(key=lambda tup: tup[1], reverse=True)
```

giving each point a specific value

```python
conjunction = set(points_1) & set(points_2)
alpha_values = np.linspace(0,1,len(conjunction))
colors = []
j = 0
for i in points:
    if i in conjunction:
        new_color = color_blending_operator.porter_duff_source_over(rot,gruen,alpha_values[j])
        colors.append(new_color)
        j+=1
    elif i[0]<=4 or i[1] <= 4:
        colors.append(np.append(rot,1))
    else:
        colors.append(np.append(gruen,1))
plt.scatter(*zip(*points),c=colors)
plt.show()
```
