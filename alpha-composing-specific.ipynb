```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import itertools
```

```python
from src import color_blending_operator
```

```python
points_1 = list(itertools.permutations(range(0,11),2))
points_2 = list(itertools.permutations(range(5,16),2))
points = list(set(points_1+points_2))
```

```python
rot = np.array([1,0,0],dtype=float)
gruen =  np.array([0,1,0],dtype=float)
```

```python
conjunction = set(points_1) & set(points_2)
alpha_values_r = np.linspace(0,1,len(conjunction))
alpha_values_g = np.linspace(1,0,len(conjunction))
```

```python
colors = []
j = 0
for i in points:
    if i in conjunction:
        new_color = color_blending_operator.alpha_composing_specific(rot,alpha_values_r[j],gruen,alpha_values_g[j])
        colors.append(new_color)
        j+=1
    elif i[0]<=4 or i[1] <= 4:
        colors.append(np.append(rot,1))
    else:
        colors.append(np.append(gruen,1))
```

```python
plt.scatter(*zip(*points),c=colors)
plt.show()
```
