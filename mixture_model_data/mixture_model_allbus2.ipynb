```python
def plot_all_methods(distributions, *args, **kwargs):
    picture_plot.plot_images(distributions, contours=True, contour_lines=True, *args, **kwargs)
    picture_plot.plot_images(distributions, pie_charts=True, *args, **kwargs)
    picture_plot.plot_images(distributions, crosses=True, contour_lines=True, *args, **kwargs)
```

```python
def generate_gaussian_plots(data, columns, n_components):
    models = []
    xlabels = []
    ylabels = []
    for i,j in itertools.combinations(columns, 2):
        models.append(MixtureModel.MixtureModel(data[[i, j]], n_components=n_components).get_gaussians())
        xlabels.append(i)
        ylabels.append(j)
    return models, xlabels, ylabels
```

```python
dataset = []
for i in range(2,5):
    models, xlabels, ylabels = generate_gaussian_plots(model_test.data, ["age", "education","income","happiness","health"], n_components=i)
    dataset.append([models, xlabels, ylabels])
```

```python
plot_all_methods(dataset[0][0], xlabels=dataset[0][1], ylabels=dataset[0][2], contour_line_level=3)
```
